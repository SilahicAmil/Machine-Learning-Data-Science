```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
ppp = pd.read_csv("PPP_data_150k_plus.csv")
```

```python
ppp
```

```python
plt.figure(figsize=(30,15))
# plt.title('Jobs Retained')
pppdata = ppp.pivot_table(values='JobsRetained',index='BusinessType',columns='State')
# sns.heatmap(pppbid)
# pppdata = ppp.pivot_table(values="NAICSCode",index="JobsRetained",columns="BusinessType")
pppData = pppdata.dropna()
sns.set_context('paper',font_scale=2)
sns.barplot(data=pppData,palette='coolwarm') 
plt.ylabel("Jobs Retained")
plt.xticks(rotation=30)
plt.tight_layout()
```

```python
sns.set_theme(style="ticks")
mpg = pd.read_csv("PPP_data_150k_plus.csv")

colors = (250, 70, 50), (350, 70, 50)
cmap = sns.blend_palette(colors, input="husl", as_cmap=True)
sns.displot(
    mpg,
    x="State",
    kind="ecdf", aspect=.75, linewidth=2, palette=cmap,
)
```
